```python
%matplotlib inline
import re
import nltk
from datasets import load_dataset
from ipywidgets import IntProgress
import pandas as pd
nltk.download('punkt',quiet=True)
nltk.download('stopwords',quiet=True)
nltk.download('wordnet',quiet=True)
nltk.download('omw-1.4',quiet=True)
nltk.download('averaged_perceptron_tagger',quiet=True)
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
from textblob import TextBlob
import matplotlib.pyplot as plt
import collections
```

```python
# Sample Data 
data = load_dataset("banking77",split="train").to_pandas()

query_column = "text"   # Column with the queries
label_column = "label"  # Column with intent names

########################
# ADD CAZOO DATA HERE
# load in the csv (or via any other format) into a pandas dataframe
# Set the "query_column" variable to the name of the column with the query text
# Set the "label_column" variable to the name of the column with intent names
# Then comment out the lines above
########################
#
# data = pd.read_csv("")
#
# query_column = "text"   # Column with the queries
# label_column = "label"  # Column with intent names
#

data.head()
```

```python
df = pd.DataFrame({"text":data[query_column],"label":data[label_column]})
df.head()
```

## Clean Text from special characters

```python
# Removes all special characters and numericals leaving the alphabets using regex
def clean(text):
    text = re.sub('[^A-Za-z]+', ' ', text)
    return text

# Cleaning the text in the review column
df['cleaned_text'] = df['text'].apply(clean)
df.head()
```

## POS Tagging (verbs, nouns, adjectives, adverbs)

Part of speech tagging to strip the word later to nothing but the root words

```python
# POS tagger dictionary
# This tags each word whether its a verb, noun, adjective or adverb
pos_dict = {'J':wordnet.ADJ, 'V':wordnet.VERB, 'N':wordnet.NOUN, 'R':wordnet.ADV}

def token_stop_pos(text):
    tags = pos_tag(word_tokenize(text))
    newlist = []
    for word, tag in tags:
        if word.lower() not in set(stopwords.words('english')):
            newlist.append(tuple([word, pos_dict.get(tag[0])]))
    return newlist

df['tagged_labels'] = df['cleaned_text'].apply(token_stop_pos)
df.head()
```

## Lemmatization 
Standardising words to "root word" to be used for sentiment analysis

```python
wordnet_lemmatizer = WordNetLemmatizer()

# Cuts out the unneeded part of the word like "ing", "ed" or "s"
def lemmatize(pos_data):
    lemma_rew = " "
    for word, pos in pos_data:
        if not pos:
            lemma = word
            lemma_rew = lemma_rew + " " + lemma
        else:
            lemma = wordnet_lemmatizer.lemmatize(word, pos=pos)
            lemma_rew = lemma_rew + " " + lemma
    return lemma_rew

df['lemma_text'] = df['tagged_labels'].apply(lemmatize)

# Finalised Cleaned data ready for sen analysis
cleaned_data = pd.DataFrame(df[["label","text","lemma_text"]])
cleaned_data.head()
```

## Sentiment Analysis

```python
# Neutral sensitivity parameter
NEUTRAL_SENS = 0.0

# function to calculate polarity
def getPolarity(review):
    return TextBlob(review).sentiment.polarity

# function to analyze the reviews
def analysis(score):
    if score < -NEUTRAL_SENS:
        return 'Negative'
    elif score >= -NEUTRAL_SENS and score <= NEUTRAL_SENS:
        return 'Neutral'
    elif score > NEUTRAL_SENS:
        return 'Positive'
```

```python
cleaned_data[:20]
```

```python
cleaned_data['polarity'] = cleaned_data['lemma_text'].apply(getPolarity) 
cleaned_data['analysis'] = cleaned_data['polarity'].apply(analysis)

neutral_text = list(cleaned_data.polarity).count(0.000000) / len(cleaned_data)

print(f"{round(neutral_text*100,5)}% of the dataset has a sentiment value of 0.")
```

```python
# Just to see some people complaining and to sanity check if they actually are complaining.
cleaned_data.sort_values("polarity")[:15]
```

## Trim the labels

```python
final_output = cleaned_data[["label","analysis"]]
final_output = final_output.groupby(['label']).analysis.value_counts(normalize=True).unstack().fillna(0)
final_output[:15]
```

## Visualisation in Seaborn

```python
import seaborn as sns
```

```python
# Get top 10 items
pos_top10 = final_output.sort_values("Positive",ascending=False)[:12]
neg_top10 = final_output.sort_values("Negative",ascending=False)[:12]
pos_top10
```

```python
pos_melt = pos_top10.reset_index().melt(id_vars=["label"])
neg_melt = neg_top10.reset_index().melt(id_vars=["label"])
pos_melt["value"] = pos_melt["value"].apply(lambda x:x*100)
neg_melt["value"] = neg_melt["value"].apply(lambda x:x*100)
pos_melt["label"] = pos_melt["label"].apply(lambda x:str(x))
neg_melt["label"] = neg_melt["label"].apply(lambda x:str(x))
pos_melt
```

```python
title = "Banking Queries Intent Analysis"
ylabel = "Intent Type"
xlabel = "Sentiment (%)"

sns.set_theme(style="whitegrid")

ax_pos = sns.catplot(x="value", y="label",legend=False, hue="analysis", kind="bar", palette=["#ea8a8a", "#ABC9EA", "#98DAA7"] ,data=pos_melt)
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
ax_neg = sns.catplot(x="value", y="label",legend=False, hue="analysis", kind="bar",palette=["#ea8a8a", "#ABC9EA", "#98DAA7"] ,data=neg_melt)
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)

ax_pos.fig.suptitle(title+" (Positive)",y=1.03,x=0.5,weight="bold",ha="center")
ax_neg.fig.suptitle(title+" (Negative)",y=1.03,x=0.5,weight="bold",ha="center")

ax_pos.set_xlabels(xlabel,weight="bold")
ax_neg.set_xlabels(xlabel,weight="bold")

ax_pos.set_ylabels(ylabel,weight="bold")
ax_neg.set_ylabels(ylabel,weight="bold")

ax_pos.fig.set_size_inches(17, 7)
ax_neg.fig.set_size_inches(17, 7)

ax_pos.savefig("sentiment_banking77_positive.png",format="png",dpi=300)
ax_neg.savefig("sentiment_banking77_negative.png",format="png",dpi=300)
```

```python
def get_n_common_words(dataframe,n,indexes):
    value_counts = {}
    
    for index in indexes:
        text_string = list(dataframe[dataframe["label"] == index]["lemma_text"])
        text_list = "".join(text_string).split()

        unique_words = set(text_list)

        freq = {}

        for word in unique_words:
            freq[word] = text_list.count(word)
        #return freq
        value_counts[index] = ", ".join(sorted(freq, key=freq.get, reverse=True)[1:n])
    return value_counts
    
get_n_common_words(cleaned_data,10,neg_top10.index)

```
